## Handle Missing Values
    *	Fill missing numerical values with the mean
    *	Fill missing categorical values with the most frequent value (mode)



```python
# 1. Handle Missing Values
# Fill missing numerical values with the mean
df['مدة الفيلم (دقيقة)'] = pd.to_numeric(df['مدة الفيلم (دقيقة)'], errors='coerce')
df['مدة الفيلم (دقيقة)'].fillna(df['مدة الفيلم (دقيقة)'].mean(), inplace=True)
```

```python
# Fill missing categorical values with the most frequent value (mode)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

## Normalize Variables
o	Normalize the  'مدة الفيلم (دقيقة)' column using Min-Max Scaling

```python
# 2. Normalize or Transform Variables
# Normalize the 'مدة الفيلم (دقيقة)' column using Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

scaler = MinMaxScaler()
df['مدة الفيلم (دقيقة)'] = scaler.fit_transform(df[['مدة الفيلم (دقيقة)']])
```

```python
output_file_path = "cleaned_cima_dataset.xlsx"
df.to_excel(output_file_path, index=False)

# Print summary of preprocessing
print("Preprocessing Completed:")
print(f"Missing values after cleaning:\n{df.isnull().sum()}")
print(f"\nSaved cleaned dataset to {output_file_path}")
```
